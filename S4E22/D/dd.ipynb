```python
import pandas as pd

# خواندن فایل CSV
file_path = 'x.csv'  # مسیر فایل CSV شما
df = pd.read_csv(file_path)

# لیست ستون‌هایی که می‌خواهیم حذف کنیم
columns_to_drop = [
    'approxdate', 'resolution', 
    'targtype2', 'targtype2_txt', 'targsubtype2', 'targsubtype2_txt', 'corp2', 'target2', 'natlty2', 'natlty2_txt',
    'targtype3', 'targtype3_txt', 'targsubtype3', 'targsubtype3_txt', 'corp3', 'target3', 'natlty3', 'natlty3_txt',
    'weaptype2', 'weaptype2_txt', 'weapsubtype2', 'weapsubtype2_txt', 
    'weaptype3', 'weaptype3_txt', 'weapsubtype3', 'weapsubtype3_txt',
    'weaptype4', 'weaptype4_txt', 'weapsubtype4', 'weapsubtype4_txt',
    'gname2', 'gsubname2', 'gname3', 'gsubname3', 
    'guncertain2', 'guncertain3', 
    'claim2', 'claimmode2', 'claimmode2_txt', 'claim3', 'claimmode3', 'claimmode3_txt', 'compclaim',
    'nhostkidus', 'ransomamtus', 'ransompaidus', 
    'nhours', 'ndays', 'divert', 'ransomnote',
    'scite2', 'scite3', 'ransomamt', 'ransompaid', 'propcomment'
]

# حذف ستون‌های مشخص‌شده
df = df.drop(columns=columns_to_drop, errors='ignore')  # errors='ignore' برای جلوگیری از خطا در صورت نبود ستون

# ذخیره دیتاست جدید در یک فایل CSV
df.to_csv('x_cleaned.csv', index=False)

# نمایش ستون‌های باقی‌مانده
print("ستون‌های باقی‌مانده:")
print(df.columns.tolist())
```

```python
from IPython.core.display import display, HTML

# محاسبه و مرتب‌سازی تعداد nullها (به ترتیب نزولی)
null_counts = df.isna().sum()
null_counts = null_counts[null_counts > 1].sort_values(ascending=False)

# ساخت جدول HTML
html_table = """
<style>
    table {
        border-collapse: collapse;
        width: 50%;
        margin: 20px 0;
        font-family: Arial;
    }
    th, td {
        border: 1px solid #888;
        padding: 8px 12px;
        text-align: left;
    }
    th {
        background-color: #222;
        color: #fff;
    }
    tr:nth-child(even) {
        background-color: #f2f2f2;
    }
</style>

<table>
    <tr><th>نام ستون</th><th>تعداد Null</th></tr>
"""

# اضافه کردن ردیف‌ها فقط برای ستون‌هایی که null > 1 دارند و مرتب شده‌اند
for col, count in null_counts.items():
    html_table += f"<tr><td>{col}</td><td>{count}</td></tr>"

html_table += "</table>"

display(HTML(html_table))
```
