```python
import os
import shutil
import json
import subprocess

dataset_path = "my_kaggle_dataset"
file_name = "abolfazl_chaanbaary.ipynb"
dataset_id = "abolfazlchanbary81/abolfazl-notebook-dataset"
metadata_file = os.path.join(dataset_path, "dataset-metadata.json")

# ساخت پوشه دیتاست
os.makedirs(dataset_path, exist_ok=True)

# کپی فایل نوت‌بوک به پوشه
if not os.path.exists(os.path.join(dataset_path, file_name)):
    shutil.copy(file_name, dataset_path)

# ساخت فایل metadata
metadata = {
    "title": "abolfazl-notebook-dataset",
    "id": dataset_id,
    "licenses": [{"name": "CC0-1.0"}]
}
with open(metadata_file, "w") as f:
    json.dump(metadata, f, indent=4)

# تابع اجرای دستور CLI
def run_command(cmd):
    process = subprocess.Popen(cmd, stdout=subprocess.PIPE, stderr=subprocess.STDOUT, text=True)
    for line in process.stdout:
        print(line, end="")
    process.wait()
    return process.returncode

# فقط دستور version اجرا کن چون create با -m پشتیبانی نمی‌شه
print("\nUploading dataset...")
ret = run_command([
    "kaggle", "datasets", "version",
    "-p", dataset_path,
    "-m", "Update notebook"
])

if ret == 0:
    print("\n✅ آپلود با موفقیت انجام شد!")
else:
    print("\n❌ خطا در آپلود!")
```
