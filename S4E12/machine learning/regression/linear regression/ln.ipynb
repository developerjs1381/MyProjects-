```python
from  sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn import metrics
import pandas as pd 
import matplotlib.pyplot as plt
import numpy as np
```

```python
df=pd.read_csv("student_scores.csv")
```

```python
df.columns
```

```python
X=df["Hours"].values.reshape(-1,1)
Y=df["Scores"].values.reshape(-1,1)
```

```python
x_train,x_test,y_train,y_test=train_test_split(X,Y,test_size=0.2,random_state=0)
```

```python
regressor=LinearRegression()
regressor.fit(x_train,y_train)
```

```python
y_pred=regressor.predict(x_test)
```

```python
plt.scatter(x_test,y_test,color="red")
plt.scatter(x_test,y_pred,color="blue")
plt.plot(x_test,y_pred)
```

```python
print("mae",metrics.mean_absolute_error(y_test,y_pred))
print("mse",metrics.mean_squared_error(y_test,y_pred))
print("rmee",np.sqrt(metrics.mean_squared_error(y_test,y_pred)))
print("R2",metrics.r2_score(y_test,y_pred))
mean_y = np.mean(y_train)
mae_ratio=metrics.mean_absolute_error(y_test,y_pred)
ggg = mae_ratio / mean_y
print(f"نسبت MAE به میانگین y: {ggg*100:.2f}%")
```

```python
import numpy as np
import matplotlib.pyplot as plt

# تنظیم x برای رسم انواع توابع
x = np.linspace(-10, 10, 400)
x_pos = np.linspace(0.1, 10, 400)  # برای لگاریتمی و نمایی

# تعریف y برای هر نوع رابطه
linear = x
quadratic = -0.5 * x**2 + 5
cubic = 0.1 * x**3 - x
logarithmic = np.log(x_pos)
exponential = np.exp(0.3 * x_pos)
sinusoidal = np.sin(x)
step = np.piecewise(x, [x < -2, (x >= -2) & (x < 3), x >= 3], [1, 3, 6])
saturating = 10 * (1 - np.exp(-0.5 * x_pos))
gaussian = np.exp(-0.5 * ((x - 0)/2)**2)

# لیست توابع برای رسم
functions = [
    ("Linear (خطی)", x, linear),
    ("Quadratic (سهمی/درجه ۲)", x, quadratic),
    ("Cubic (درجه ۳)", x, cubic),
    ("Logarithmic (لگاریتمی)", x_pos, logarithmic),
    ("Exponential (نمایی)", x_pos, exponential),
    ("Sinusoidal (نوسانی)", x, sinusoidal),
    ("Step (پله‌ای)", x, step),
    ("Saturating (اشباع‌شونده)", x_pos, saturating),
    ("Gaussian (زنگوله‌ای)", x, gaussian),
]

# رسم نمودارها
fig, axs = plt.subplots(3, 3, figsize=(15, 12))
axs = axs.flatten()

for i, (title, x_vals, y_vals) in enumerate(functions):
    axs[i].plot(x_vals, y_vals, label=title, color='teal')
    axs[i].set_title(title)
    axs[i].grid(True)

plt.tight_layout()
plt.show()
```

```python
from flask import Flask, request, render_template_string
import threading
import mysql.connector

app = Flask(__name__)

# HTML صفحه اصلی
html_template = '''
<!doctype html>
<html lang="fa">
<head>
  <meta charset="UTF-8">
  <title>مدل ماشین لرنینگ</title>
  <style>
    @import url('https://fonts.googleapis.com/css2?family=Vazirmatn:wght@400;700&display=swap');

    body {
      font-family: 'Vazirmatn', sans-serif;
      background: linear-gradient(135deg, #0f172a, #1e293b);
      min-height: 100vh;
      margin: 0;
      padding: 0;
      display: flex;
      flex-direction: column;
      align-items: center;
      justify-content: center;
      color: #e2e8f0;
      direction: rtl;
    }

    h2 {
      font-size: 36px;
      margin-bottom: 20px;
      color: #60a5fa;
    }

    form {
      background: rgba(255, 255, 255, 0.05);
      backdrop-filter: blur(12px);
      border-radius: 20px;
      box-shadow: 0 8px 20px rgba(0, 0, 0, 0.3);
      padding: 40px 35px;
      width: 100%;
      max-width: 480px;
      display: flex;
      flex-direction: column;
      gap: 20px;
    }

    label {
      font-size: 18px;
      margin-bottom: 5px;
    }

    input[type="number"] {
      width: 70%;
      padding: 14px;
      background-color: #1e293b;
      color: #e2e8f0;
      border: 2px solid #334155;
      border-radius: 12px;
      font-size: 18px;
      transition: border 0.3s;
    }

    input[type="number"]:focus {
      border-color: #60a5fa;
      outline: none;
    }

    input[type="submit"] {
      padding: 14px;
      background: #3b82f6;
      color: white;
      border: none;
      border-radius: 12px;
      font-size: 20px;
      font-weight: bold;
      cursor: pointer;
      transition: background 0.3s;
    }

    input[type="submit"]:hover {
      background: #2563eb;
    }

    .result, .error {
      margin-top: 30px;
      padding: 20px;
      border-radius: 16px;
      width: 100%;
      max-width: 480px;
      font-size: 18px;
    }

    .result {
      background: rgba(56, 189, 248, 0.15);
      color: #7dd3fc;
      border: 1px solid #38bdf8;
    }

    .error {
      background: rgba(248, 113, 113, 0.15);
      color: #f87171;
      border: 1px solid #ef4444;
    }

    a {
      margin-top: 30px;
      text-decoration: none;
      color: #93c5fd;
      font-size: 18px;
      font-weight: bold;
      transition: color 0.3s;
    }

    a:hover {
      color: #60a5fa;
    }
  </style>
</head>
<body>
  <h2>مقادیر را وارد کنید:</h2>
  
  <form method="POST">
    <div>
      <label for="h">Hours</label>
      <input type="number" step="any" name="h" id="h" required>
    </div>
    <div>
      <label for="t">Practice</label>
      <input type="number" step="any" name="t" id="t" required>
    </div>
    <div>
      <label for="p">TeamWork</label>
      <input type="number" step="any" name="p" id="p" required>
    </div>
    <input type="submit" value="محاسبه و ذخیره">
  </form>

  {% if y is not none %}
    <div class="result">
      <h3>نتیجه y: {{ y }}</h3>
      <p>مقدارهای واردشده: h = {{ h }}, t = {{ t }}, p = {{ p }}</p>
    </div>
  {% endif %}

  {% if error %}
    <div class="error">
      <h3>خطا: {{ error }}</h3>
    </div>
  {% endif %}

  <a href="/dataset">نمایش دیتاست</a>
</body>
</html>


'''

# تابع اتصال به دیتابیس
def get_db_connection():
    return mysql.connector.connect(
        host="localhost",
        user="developer",
        password="",  # پسورد دیتابیس خودت رو اینجا بذار
        database="love"
    )

# صفحه اصلی
@app.route('/', methods=['GET', 'POST'])
def index():
    y = None
    error = None
    h = t = p = None

    if request.method == 'POST':
        try:
            h = float(request.form.get('h'))
            t = float(request.form.get('t'))
            p = float(request.form.get('p'))

            # مدل ماشین لرنینگ (رگرسیون خطی)
            y = 8.37366712 + (6.77781271 * h) + (-0.03200033 * t) + (5.07424529 * p)

            # ذخیره در دیتابیس
            conn = get_db_connection()
            cursor = conn.cursor()
            cursor.execute(
                "INSERT INTO ali (h, t, p, y) VALUES (%s, %s, %s, %s)",
                (h, t, p, y)
            )
            conn.commit()
            cursor.close()
            conn.close()
        except Exception as e:
            error = str(e)
            print("Error:", error)

    return render_template_string(html_template, y=y, error=error, h=h, t=t, p=p)

# صفحه نمایش دیتاست
@app.route('/dataset')
def show_dataset():
    conn = get_db_connection()
    cursor = conn.cursor()
    cursor.execute("SELECT h, t, p, y FROM ali")
    data = cursor.fetchall()
    cursor.close()
    conn.close()

    table_html = '''
    <h2>دیتاست ثبت‌شده</h2>
    <table border="1" cellpadding="5">
      <tr>
        <th>h</th><th>t</th><th>p</th><th>y</th>
      </tr>
      {% for row in data %}
        <tr>
          <td>{{ row[0] }}</td>
          <td>{{ row[1] }}</td>
          <td>{{ row[2] }}</td>
          <td>{{ row[3] }}</td>
        </tr>
      {% endfor %}
    </table>
    <br><a href="/">برگشت به فرم</a>
    '''
    return render_template_string(table_html, data=data)

# اجرای برنامه روی پورت 8001
def run_app():
    app.run(port=8016, debug=False, use_reloader=False)

threading.Thread(target=run_app).start()
```

```python
import numpy as np
import pandas as pd
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
import matplotlib.pyplot as plt

# خواندن داده‌ها
df = pd.read_csv("Data.csv")

# ساخت فیچرها با توان‌های بالا برای ایجاد overfitting
X_custom = pd.DataFrame({
    'Hours_1': df['Hours'],
    'Hours_2': df['Hours']**2,
    'Practice_1': df['Practice'],
    'Practice_2': df['Practice']**2,
    'TeamWork_1': df['TeamWork'],
    'TeamWork_2': df['TeamWork']**2,

})

y = df['Scores']

# تقسیم‌بندی داده‌ها
X_train, X_test, y_train, y_test = train_test_split(X_custom, y, test_size=0.3, random_state=42)

# مدل
model = LinearRegression()
model.fit(X_train, y_train)

# پیش‌بینی و نتایج
print("عرض از مبدا:", model.intercept_)
print("\nوزن ویژگی‌ها:")
for name, coef in zip(X_custom.columns, model.coef_):
    print(f"{name}: {coef:.4f}")

# پیش‌بینی برای رسم نمودار
y_pred_test = model.predict(X_test)

# مرتب‌سازی بر اساس Hours برای تست
sorted_indices_test = X_test['Hours_1'].argsort()
hours_sorted_test = X_test['Hours_1'].iloc[sorted_indices_test]
y_test_sorted = y_test.iloc[sorted_indices_test]
y_pred_sorted_test = y_pred_test[sorted_indices_test]

plt.figure(figsize=(10,6))
plt.scatter(hours_sorted_test, y_test_sorted, color='blue', label='Real Test Data')
plt.plot(hours_sorted_test, y_pred_sorted_test, color='red', label='Prediction on Test Data')
plt.xlabel('Hours')
plt.ylabel('Scores')
plt.legend()
plt.title('Model Prediction on Test Data')
plt.grid(True)
plt.tight_layout()
plt.show()
```
